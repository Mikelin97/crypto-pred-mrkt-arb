```python
import numpy as np
import pandas as pd
```

```python
from rpy2.robjects import pandas2ri, r
import rpy2.robjects.packages as rpackages
```

```python
import rpy2.robjects as ro
from rpy2.robjects.conversion import localconverter
from rpy2.robjects import pandas2ri
import rpy2.robjects.packages as rpackages
from rpy2.robjects import default_converter
# Load the R package

with localconverter(default_converter + pandas2ri.converter):
    # Importing R packages inside this block ensures conversions are initialized
    highfrequency = rpackages.importr("highfrequency")
```

```python
timestamps = pd.date_range("2025-01-01 09:30", periods=60, freq="1min")
prices = 100 + np.cumsum(np.random.randn(len(timestamps)) * 0.1)
data = pd.DataFrame({"PRICE": prices}, index=timestamps)

# Convert DataFrame → R object safely
with localconverter(default_converter + pandas2ri.converter):
    r_data = ro.conversion.py2rpy(data)

# Compute realized covariance
r_result = highfrequency.rCov(r_data)

# Convert R → Python safely
with localconverter(default_converter + pandas2ri.converter):
    result_py = ro.conversion.rpy2py(r_result)

print("Realized variance result:")
print(result_py)
```
