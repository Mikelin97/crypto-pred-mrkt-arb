### Populating Series

```python
from typing import List, Optional
from pydantic import BaseModel
from datetime import datetime
```

```python
import requests

params = {
    "limit": 100,
    "offset": 0
}
url = "https://gamma-api.polymarket.com/series"

all_series = []

response = requests.get(url, params=params)
new_series = response.json()
while new_series:
    all_series += new_series
    params['offset'] += 100
    response = requests.get(url, params=params)
    new_series = response.json()


# print(response.json())
```

```python
len(all_series)
```

```python
class Series(BaseModel):
    series_id: str
    ticker: str
    slug: str
    title: Optional[str] = None
    series_type: Optional[str]
    recurrence: Optional[str]
    active: Optional[bool]
    closed: Optional[bool]
    published_at: Optional[datetime]
    updated_at: Optional[datetime]
    start_date: Optional[datetime]
    created_at: Optional[datetime]
```

```python
series = all_series[0]
```

```python
series_objects = []
```

```python
for series in all_series:

    if 'startDate' in series:
        print(series['startDate'])
    published_at = series.get('publishedAt')
    if isinstance(published_at, str):
        published_at += ":00"
    clean_series = Series(
            series_id=series.get('id'),
            ticker=series.get('ticker'),
            slug=series.get('slug'),
            title=series.get('title'),
            series_type=series.get('seriesType'),
            recurrence=series.get('recurrence'),
            active=series.get('active'),
            closed=series.get('closed'),
            published_at=published_at, # very sparse
            updated_at=series.get('updatedAt'),
            start_date=series.get('startDate'), # very sparse
            created_at=series.get('createdAt'),
        )
    
    series_objects.append(clean_series)
    
```
