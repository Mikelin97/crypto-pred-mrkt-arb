```python
import numpy as np
from scipy.stats import norm
```

```python
# d1 = (np.log(S/X) + (b + (vol**2)/2) * T) / (vol * np.sqrt(T))
# d2 = d1 - vol * np.sqrt(T)
def binary_price(S, X, T, vol, r):
    d2 = (np.log(S/X) - (r - (vol**2)/2) * T) / (vol * np.sqrt(T))

    price = np.exp(-r*T) * norm.cdf(d2)
    return price
```

```python
S = 113580
X = 113709.59
r = 0.0419515 # sofr
min_term = 15
secs = 0
seconds_to_expiry = min_term*60 + secs
time_to_expiry = (seconds_to_expiry / 86400) / 365 # in years
vol = 0.43 # volatility of our crypto

binary_price(S, X, time_to_expiry, vol, r)
```
